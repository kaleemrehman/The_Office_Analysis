# Question# 3 Average number of words for each characters

```python
df['line_text'] = df['line_text'].apply(str.lower)
df['line_text'] = df['line_text'].replace('\d+', '', regex = True)
df['line_text'] = df['line_text'].replace('[^\w\s\+]', '', regex = True)
```

```python
df['word_count']=df["line_text"].apply(lambda x: len(str(x).split(' ')))
```

```python
df
```

```python
df2=df.groupby('speaker')['word_count'].mean()
```

```python
df2
```

```python
df2.nlargest(10)
```
